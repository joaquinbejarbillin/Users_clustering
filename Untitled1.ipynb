```python
invoices = db.get_collection('invoices')
```

```python
pipeline = [
    {
      "$unwind": {
        "path": '$historical',
        "preserveNullAndEmptyArrays": True,
      },
    },

    {
      "$sort": {
        'historical.updated_at': 1,
      },
    },
    {
      "$match": {
        "$or": [
          { "historical": { "$exists": "false" } },
          {
            "$and": [
              { 'historical': { '$ne': None } },
              { 'historical.new_status': { '$ne': None } },
              { 'historical.updated_at': { '$ne': None } },
            ],
          },
        ],
      },
    },

    {
      '$group': {
        '_id': {
          '_id': '$_id',
          'updated_at': '$updated_at',
          'created_at': '$created_at',

          'from_recurrent': '$from_recurrent',
        },
        'historical': { '$push': '$historical' },
      },
    },

    {
      '$project': {
        '_id': 1,
        'historical': {
          '$arrayToObject': {
            '$map': {
              'input': '$historical',
              'as': 'pair',
              'in': ['$$pair.new_status', '$$pair.updated_at'],
            },
          },
        },
      },
    },

    {
      '$project': {
        'historical.issued': { '$ifNull': ['$historical.issued', '$historical.sent'] },
        'historical.sent': { '$ifNull': ['$historical.sent', '$historical.sent'] },
        'historical.viewed': { '$ifNull': ['$historical.viewed', '$historical.viewed'] },
      },
    }]
```

```python
j = invoices.aggregate(pipeline=pipeline, allowDiskUse=True)

```

```python
df =  pd.DataFrame(list(j))
```
